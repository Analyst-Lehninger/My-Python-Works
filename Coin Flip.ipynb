```python
def coin_flip(): #define a binary outcome with random 0,1
  if random.random() >= 0.5:
    return "Heads"
  else:
    return "Tails"
```

```python
coin_flip() #test the function return
```

```python
def n_flips(n): #fxn for n times
  heads_count = 0
  tails_count = 0
  for i in range(n):
    if coin_flip() == 'Heads':
      heads_count += 1
    if coin_flip() == "Tails":
      tails_count += 1
  print(heads_count/(heads_count+tails_count)*100, tails_count/(heads_count+tails_count)*100)
```

```python
n_flips(9999)
```
